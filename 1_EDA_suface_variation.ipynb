# EDA 
Exploratory data analysis of the images for the study of fire correlation with surface type from a little surface of the Amazonian Forest.

All the images have been extracted from Google Earth Engine.   
The code used is in the file ESA_FireCCI.js

The images are taken from the dataset : FireCCI51: MODIS Fire_cci Burned Area Pixel Product, Version 5.1
You can find more information about it in:   https://developers.google.com/earth-engine/datasets/catalog/ESA_CCI_FireCCI_5_1#bands    
the images have a resolution of 250m per pixel.


In this study, I chose to focus on a specific area within the Amazon rainforest. 

This region is selected due to its high incidence of wildfires, which is notably high despite being one of Earth's most humid locations. Surprisingly, wildfires in this area spread more rapidly than one would expect, given the environmental conditions. 

The Amazon rainforest is a dynamic ecosystem, constantly impacted by factors like deforestation, agricultural land use, and significant rainfall. This fast changing landscape presents an opportunity to explore the relationship between wildfire occurrences and the evolving terrain over a period of two years.     

For this study, I selected a specific location in Brazil, close to the border with Bolivia. This area is characterized by an abundance of water bodies, extensive croplands, patches of forest, and areas of dense natural vegetation. The unique combination of these diverse ecological components makes it an ideal site for examining environmental interactions and changes.

```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
from matplotlib.colors import  Normalize
import numpy as np
```

```python
file_path = 'FireCCI_LandCover_GEE/ImagesGoogleEE.2020_10_01_LandCover.tif'

with rasterio.open(file_path) as src:
    data = src.read(1)
```

```python
print(f"The size of the images is: {data.shape}")
print(f"Total number of pixels to work: {data.shape[0]**2} pixels")
print(f"The real surface that we are working with: {(data.shape[0]*250/1000)**2} km²")
```

### Exemple of the an image from the date: 2020_10_01

```python
plt.figure(figsize=(10, 10))
img_plot = plt.imshow(data, cmap='viridis', norm=Normalize(vmin=0, vmax=180))

cbar = plt.colorbar(img_plot)
cbar.set_label('Value')

cbar.set_ticks(np.arange(0, 181, 10))
cbar.set_ticklabels(np.arange(0, 181, 10))

plt.title('Burned Area')
plt.show()
```

The values on the Data go from 0 to 180 in an intervalle of 10. The colors are not the same.
